```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dapo_rewards.csv', header=None)
# split the first col into two cols by comma
df[[0, 1]] = df[0].str.split(',', expand=True)

# remove letters from the first col
df[1] = df[1].str.replace('[^0-9.]', '', regex=True).astype('float')
df[0] = df[0].str.replace('[^0-9.]', '', regex=True).astype('int')
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# set all figure dpi to 300
plt.rcParams['figure.dpi'] = 300
```

```python
# plot the second column with seaborn
sns.set_style("darkgrid")
sns.lineplot(data=df[1])
# add OLS
sns.regplot(x=df[0], y=df[1], ci=None, scatter=False, color="red")
plt.xlabel('Epoch')
plt.ylabel('Total reward')
plt.show()
```
